```python
import pandas as pd

df = pd.read_csv('datasets/siena2018-pres.csv')
```

1. Sort the index.

```python
df.sort_index()
```

2. Set the index to a string column, sort the index, and slice by a substring of index values.


```python

df.set_index('Party').sort_index().loc['Democratic':'Independent']
```

3. Sort by a single column.

```python
df.sort_values(by='Party')
```

4. Sort by a single column in descending order.


```python
df.sort_values(by='Party', ascending=False)
```

5. Sort by two columns.


```python
df.sort_values(by=['Party', 'President'])
```

6. Sort by the last letter of a string column.


```python
df.assign(tmp=df["Party"].str.slice(-1)).sort_values("tmp").drop("tmp", 1)
```
