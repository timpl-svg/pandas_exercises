```python
import pandas as pd

df = pd.read_csv('datasets/population_by_age_group.csv', low_memory=False)
df
```

1. Use the debugger to step into the chain of your tweak function.

```python
def tweak_population(df_):
    return (df_.rename(columns=lambda c: c.replace('-', '_'))
               .rename(columns=lambda c: c.replace('+', 'more'))
               .rename(columns=lambda c: c.replace(' ', '_')))
                

tdf = tweak_population(df)
tdf
```

2. Capture an intermediate state of your chain into a variable.

```python
class DebugException(Exception):
    pass

def debug_var(thing, *, name='debug_item', raise_ex=True):
    globals()[name] = thing
    if raise_ex:
        raise DebugException
    return thing

tdf = tweak_population(df)
tdf.apply(debug_var, name='this')
```
