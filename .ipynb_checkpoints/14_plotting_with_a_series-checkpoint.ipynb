```python
import pandas as pd

df = pd.read_csv('datasets/alta_noaa_1980_2019.csv')
dates = pd.to_datetime(df.DATE)
snow = df.SNOW.rename(dates)
```

1. Create a histogram from a numeric column. Change the bin size.

```python
snow.plot.hist(bins=20)
```

2. Create a boxplot from a numeric column.


```python
snow.plot.box()
```

3. Create a Kernel Density Estimate plot from a numeric column.


```python
snow.plot.kde()
```

4. Create a line from a numeric column.


```python
snow.plot.line()
```

5. Create a bar plot from a frequency count of a categorical column.


```python
df = pd.read_csv('datasets/vehicles.csv')
make = df.make
make.value_counts().plot.bar()
```

6. Create a pie plot from a frequency count of a categorical column.


```python
season2017 = (snow.loc['2016-10':'2017-05'])
season2017.resample('M').sum().div(season2017.sum()).mul(100).rename(lambda idx:idx.month_name()).plot.pie(title='2017 Monthly Percent of Snowfall')
```
